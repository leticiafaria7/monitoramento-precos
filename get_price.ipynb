### Obter o preço dos produtos

```python
# bibliotecas

import requests
from bs4 import BeautifulSoup
```

```python
# natura

site_amazon = 'https://www.amazon.com.br/dp/B09WTSKRTN/ref=olp-opf-redir?aod=1&ie=UTF8&condition=new&creative=380345&creativeASIN=B09WTSKRTN&linkCode=asm&tag=zoom059-20&ascsubtag=41e46cee9ffd46afbf89cf628b876e27'

response = requests.get(site_amazon)
content = response.content

site = BeautifulSoup(content, 'html.parser')

print(site.prettify())

# preco = site.find('div', attrs={'class': 'ppd'})
# print(preco)
# preco2 = preco.find('span', attrs = {'class': "a-offscreen"})
# print(preco2)
```
